```python
#  Import the required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Prepare the dataset
raw_text = "I love machine learning. It is breathtaking! Deep learning and natural language processing are pretty phenomenal."
```

```python
# Preprocess the data
chars = sorted(set(raw_text))
char_to_int = dict((c, i) for i, c in enumerate(chars))
int_to_char = dict((i, c) for i, c in enumerate(chars))
```

```python
# Create input-output sequences
sequence_length = 10
input_sequences = []
output_sequences = []

for i in range(0, len(raw_text) - sequence_length, 1):
    input_seq = raw_text[i: i + sequence_length]
    output_seq = raw_text[i + sequence_length]
    input_sequences.append([char_to_int[char] for char in input_seq])
    output_sequences.append(char_to_int[output_seq])

X = np.reshape(input_sequences, (len(input_sequences), sequence_length, 1)) / float(len(chars))
y = to_categorical(output_sequences)
```

```python
# Build the RNN model using Keras with one SimpleRNN layer and one Dense layer for output
model = Sequential()
model.add(SimpleRNN(128, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
# Train the model
model.fit(X, y, epochs=100, batch_size=32)
```

```python
# Text generation
def generate_text(seed_text, num_chars):
    generated_text = seed_text
    
    for _ in range(num_chars):
        seed_text_padded = seed_text[-sequence_length:]
        input_seq = np.array([char_to_int[char] for char in seed_text_padded]).reshape((1, sequence_length, 1)) / float(len(chars))
        prediction = model.predict(input_seq, verbose=0)
        predicted_char = int_to_char[np.argmax(prediction)]
        seed_text = seed_text[1:] + predicted_char
        generated_text += predicted_char
        
    return generated_text

seed_text = "I love mach"
generated_text = generate_text(seed_text, 50)
print("Generated text:", generated_text)
```

```python
import unittest

class TestRNNTextGeneration(unittest.TestCase):

    def test_generated_text_length(self):
        seed_text = "I love mach"
        num_chars = 50
        generated_text = generate_text(seed_text, num_chars)
        self.assertEqual(len(generated_text), len(seed_text) + num_chars)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
